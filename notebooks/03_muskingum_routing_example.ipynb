# 示例: 2. 马斯京根汇流示例

**脚本:** `examples/run_muskingum_example.py`

## 目的
演示 `MuskingumRouting` 模块的用法。脚本定义了马斯京根参数 `K` 和 `x`，然后将一个三角形的入流过程线通过该模块进行演算，最终绘制出入流和出流过程线，以展示其对洪峰的削减和延迟效果。

## 1. 环境设置

首先，我们导入所需的库，并把项目的根目录添加到Python的搜索路径中。

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# 将项目根目录添加到Python路径
sys.path.insert(0, os.path.abspath(os.path.join(os.path.dirname('__file__'), '..')))
from hydro_model.routing import MuskingumRouting
```

## 2. 定义参数

马斯京根法的核心参数是 `K` (槽蓄系数) 和 `x` (流量比重系数)。`K`反映了洪水波在河段中传播的时间，`x`则反映了河段槽蓄量与出入流量的关系。

```python
params = {'K': 2.0, 'x': 0.2} # K=2天, x=0.2
```

## 3. 创建汇流模块实例

我们使用上面定义的参数来创建一个`MuskingumRouting`的实例。

```python
muskingum_router = MuskingumRouting(**params)
```

## 4. 创建入流过程线

我们创建一个人工的三角形状的入流过程线，用来模拟一个简单的洪水波。

```python
inflow_hydrograph = np.array([0, 10, 20, 40, 60, 45, 30, 20, 10, 5, 2, 1, 0, 0, 0])
timesteps = np.arange(len(inflow_hydrograph))

print(f"输入入流过程线: {inflow_hydrograph}")
```

## 5. 运行汇流模拟

我们逐个时间步地将入流值输入到汇流模块中，并收集计算得到的出流值。

```python
outflow_hydrograph = []
for inflow_val in inflow_hydrograph:
    outflow_val = muskingum_router.run(inflow_val)
    outflow_hydrograph.append(outflow_val)

outflow_hydrograph = np.array(outflow_hydrograph)

print(f"输出出流过程线: {np.round(outflow_hydrograph, 2)}")
```

## 6. 可视化结果

最后，我们绘制入流和出流过程线，以直观地比较它们之间的差异。可以看到，经过马斯京根法演算后，洪峰被明显削减，并且出现了一定的延迟，这符合预期的物理过程。

```python
plt.figure(figsize=(10, 6))
plt.plot(timesteps, inflow_hydrograph, 'b-o', label='Inflow Hydrograph')
plt.plot(timesteps, outflow_hydrograph, 'r--s', label='Outflow Hydrograph (Routed)')
plt.title('Muskingum Routing Example')
plt.xlabel('Time Step (days)')
plt.ylabel('Flow (m³/s)')
plt.legend()
plt.grid(True)

output_dir = '../examples/results/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_path = os.path.join(output_dir, 'muskingum_example_plot.png')
plt.savefig(output_path)
print(f"\n汇流图已保存到 {output_path}")
plt.show()
```
