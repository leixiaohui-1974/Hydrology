# 欢迎 & 后续学习指南

## 欢迎!

欢迎来到本水文-水力耦合模型框架的Jupyter Notebook文档系列！如果您看到了这里，说明您已经拥有了一套包含50个文档的、从入门到精通的完整学习材料。

本系列文档旨在：
- **演示核心功能**: 通过`examples/`目录中的示例，展示框架的核心能力。
- **深入解析组件**: 对每个关键的模型组件进行深入的、独立的剖析。
- **提供实用教程**: 针对建模过程中的常见任务（如数据处理、结果分析、调试）提供“食谱式”的教程。

## 学习路径建议
对于初学者，我们建议您按照以下路径来学习，这能帮助您循序渐进地掌握整个框架：

1.  **从一个完整的例子开始**
    - 首先通读 **`01_basic_model_example.ipynb`**。它会带您完整地体验一次从数据加载、模型设置到运行和可视化的全过程，对框架建立一个初步的整体印象。

2.  **了解核心模型组件**
    - 阅读API参考文档，了解两个核心的模型类：
        - **`42_api_reference_hydrologicalmodel.ipynb`**
        - **`43_api_reference_hydraulicmodel.ipynb`**

3.  **掌握模型耦合机制**
    - 理解`SimulationController`是如何将不同的组件连接在一起的。
        - **`11_coupled_model_example.ipynb`**
        - **`20_controller_deep_dive.ipynb`**

4.  **探索高级工作流**
    - 当您掌握了基础后，可以探索框架提供的自动化高级工作流。
        - **`08_gis_delineation_workflow.ipynb`**: 学习如何处理GIS数据。
        - **`09_enkf_calibration_workflow.ipynb`**: 学习如何使用EnKF进行模型率定。
        - **`26_running_from_config_file.ipynb`**: 学习如何通过YAML文件来驱动模拟。

5.  **查阅专题和应用案例**
    - 最后，您可以根据自己的兴趣和需求，查阅“教程”和“应用案例”部分，学习特定的技巧或应用场景，例如：
        - **`30_land_use_change_impact.ipynb`**
        - **`31_flood_control_gate_impact.ipynb`**
        - **`38_result_analysis_fdc.ipynb`**
        - **`48_debugging_linalgerror.ipynb`**
