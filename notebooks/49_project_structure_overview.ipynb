# 指南: 项目结构概览

## 目的
对于任何一个新接触本项目的开发者或用户来说，首先需要了解的是整个代码库是如何组织的。本篇文档旨在提供一个清晰的项目“地图”，解释每个主要目录和文件的用途。


## 顶层目录结构

```
.
├── common/             # 核心框架与公共组件
├── data/               # 示例用的CSV数据文件
├── docs/               # 原始的Markdown格式文档
├── examples/           # 独立的Python示例脚本
├── gis_data/           # 示例用的GIS数据
├── gui/                # GUI图形用户界面
├── hydro_model/        # 概念性集总式水文模型
├── model_2d/           # 二维水动力模型 (概念验证)
├── notebooks/          # Jupyter Notebook 格式的详细文档 (即本文档系列)
├── preissmann_model/   # 一维水动力模型 (Preissmann)
├── results/            # 示例脚本默认的输出目录
└── ...                 # 其他独立的脚本和配置文件
```

## 目录详解

### `common/`
存放整个耦合模型框架的核心与公共组件。
- **`base_model.py`**: 定义了所有模型组件都必须继承的抽象基类 `BaseModelComponent`。这是实现框架模块化和可扩展性的基石。
- **`controller.py`**: 定义了“指挥中心” `SimulationController`，负责管理和运行整个模型网络。
- **`junction.py`**: 定义了用于处理河网汇流和分流的`Junction`组件。
- **`config_parser.py`**: 定义了`ConfigParser`，用于解析YAML配置文件并构建仿真。

### `hydro_model/`
存放概念性的、集总式的水文模型组件。该模型主要用于降雨-径流模拟。
- **`model.py`**: `HydrologicalModel`类，将产流和汇流模块组合在一起。
- **`runoff.py`**: 包含了不同的产流模块，如`SCSCurveNumberModule`。
- **`routing.py`**: 包含了不同的概念性汇流模块，如`MuskingumRouting`和`UnitHydrographRouting`。
- **`enkf.py`**: 实现了集合卡尔曼滤波（EnKF），用于模型校准和数据同化。

### `preissmann_model/`
存放基于Preissmann隐式格式的一维水动力模型。主要用于模拟河道中的非恒定流。
- **`model.py`**: `HydraulicModel`类，负责组装和求解圣维南方程。
- **`solver.py`**: 包含一个分块三对角矩阵的求解器（**注意：当前未被`model.py`使用**）。
- **`reach.py`**: `RiverReach`类，定义了由断面组成的河段几何。
- **`cross_section.py`**: 定义了不同的断面形状，如`RectangularCrossSection`。
- **`structures.py`**: 定义了可以放置在河道中的水工建筑物，如`Gate`和`Pump`。

### `model_2d/`
存放一个概念验证阶段的二维水动力模型。
- **`model.py`**: `Model2D`类，是二维模型的入口。
- **`mesh.py`**: 定义了非结构化三角网格所需的数据结构（`Node`, `Edge`, `Face`）。
- **`solver.py`**: `finite_volume_step`函数，实现了有限体积法求解器。

### `gui/`
存放GUI图形用户界面的所有文件。
- **`main.py`**: GUI的Python后端，使用`eel`库启动服务并暴露计算函数。
- **`web/`**: 存放所有前端文件（HTML, CSS, JavaScript）。`app.js`是前端的主要逻辑文件。

### `data/`, `gis_data/`, `examples/`
这三个目录提供了丰富的示例，是学习如何使用本框架的最佳起点。
- **`data/`**: 存放表格数据（CSV），如时间序列。
- **`gis_data/`**: 存放地理空间数据，如DEM（`.tif`）和Shapefile（`.shp`）。
- **`examples/`**: 存放一系列独立的Python脚本，每个脚本都演示了一个特定的功能或模型。

### `docs/` & `notebooks/`
这两个目录是项目主要的文档来源。
- **`docs/`**: 存放项目早期的、由开发者编写的Markdown格式的文档。
- **`notebooks/`**: 存放本系列Jupyter Notebook格式的详细教程、API参考和应用案例，提供了更系统、更具交互性的学习体验。
